## 3. ***Data Wrangling***

### Data Wrangling Code

````python

# Write your code to make your dataset analysis ready.

# Removing duplicates in app data

app_data.drop_duplicates(inplace=True)

#removing duplicates in review data

review_data.drop_duplicates(inplace=True)

# Removing null values in app data

app_data.dropna(inplace=True)

# Removing null values in review data

review_data.dropna(inplace=True)

# checking for outliers in app data(rating column)

outliers = app_data[(app_data["Rating"] < 1) | (app_data["Rating"] > 5)]         #for ratings
print (outliers)

# checking for outliers in review data(Sentiment Polarity and Sentiment Subjectivity column)

outliers = review_data[(review_data["Sentiment_Polarity"] < -1) | (review_data["Sentiment_Polarity"] > 1)]      #for sentiment polarity
print(outliers)

outliers = review_data[(review_data["Sentiment_Subjectivity"] < 0) | (review_data["Sentiment_Subjectivity"] > 1)]     #for sentiment subjectivity
print(outliers)

# Only doing outlier check on these three columns in both dataset because these columns only can have maximum outliers as they are numeric columns with fixed range. Therefore, no outliers found.

# Removing + sign from installs so that it can be used as numerical value

app_data["Installs"] = app_data["Installs"].str.replace("[+,]", "", regex=True).astype(int)

# Data type conversion (app data)

app_data['Last Updated'] = pd.to_datetime(app_data['Last Updated'])     # converting object to datetime because it makes data sorting easier

app_data['Reviews'] = app_data['Reviews'].astype(int)        # converting object to int because review column has numeric values

app_data['Installs'] = app_data['Installs'].astype(int)      ## converting object to int because installs column has numeric values

# Mering both datasets for further visualizations

merged_df = pd.merge(app_data, review_data, on="App", how="inner")

````

````python
# checking the data wrangling operations

print("Duplicate items:" , merged_df.duplicated().sum())     # for duplicates

print("Null values:" , merged_df.isnull().sum())             # for null values

merged_df.info()                                             # for data type conversion

merged_df.head()                                             # first look of merged dataset

````
